```
import pandas as pd
```

https://docs.google.com/spreadsheets/d/186f9JePfK6rlRrHCcf0aD-lqEQEXCkgJlYhKlXdjZgE/edit?usp=sharing

```
url = 'https://docs.google.com/spreadsheets/d/186f9JePfK6rlRrHCcf0aD-lqEQEXCkgJlYhKlXdjZgE/edit?usp=sharing'
```

```
id = url.split('/')[5]
```

```
df = pd.read_csv(f'https://docs.google.com/spreadsheets/d/{id}/export?format=csv')
```

```
df
```

```
df.groupby('Housing')['default'].agg(['count','mean'])
```

```
df['amount_bin'] = pd.cut(df['Credit amount'],5)
```

```
df.groupby('amount_bin')['default'].agg(['count','mean'])
```

```
df['amount_bin_q'] = pd.qcut(df['Credit amount'],10)
```

```
df.groupby('amount_bin_q')['default'].agg(['count','mean'])
```

```
df.groupby('amount_bin_q')['default'].mean().plot(ylim=(0,1))
```

```
df.groupby('amount_bin_q')['default'].mean().plot(kind='bar')
```

```
df['Credit amount'].hist()
```

```
df.boxplot('Credit amount')
```

```
df.boxplot('Credit amount',by='Housing')
```

```
df.hist('Credit amount',by='Housing')
```

```
t = df.groupby('Housing')['Credit amount'].agg(['count','median'])
```

```
t
```

```
ax1 = t['median'].plot(ylim=(0,5000))
ax2 = ax1.twinx()
t['count'].plot(ax=ax2,kind='bar',alpha=0.2)
```

```
df.plot(x='Age',y='Credit amount',kind='scatter')
```

```
df['age_bin'] = pd.cut(df['Age'],5)
```

```
t = df.groupby('age_bin')['Credit amount'].agg(['count','median'])
```

```
t
```

```
ax1 = t['median'].plot(ylim=(0,5000))
ax2 = ax1.twinx()
t['count'].plot(ax=ax2,kind='bar',alpha=0.2,rot=45)
```
