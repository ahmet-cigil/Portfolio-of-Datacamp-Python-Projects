```python
# The above error is because the series is seriesgroupby object and needs apply method
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

### **What is the percentage of programmers who know Python per country?**

```python
# 1.Method 
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
```

```python
python_df
```

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents']) * 100
python_df
```

```python
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
python_df
```

```python
python_df.loc['Netherlands']
```

```python
# 2.Method
ctr_knows_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python', na=False).value_counts(normalize=True))
ctr_knows_python.rename({False:'Don\'t know', True:'I know'}, inplace=True)
ctr_knows_python
```

## Select from PostgreSQL  
I repeat some imports and variables here because I usually copy this cell to the beging of a new project

```python
import psycopg2 as pg2
import pandas as pd

con = pg2.connect(host='this_postgres',
                  user='postgres',
                  password='password',
                  database='postgres')
con.autocommit = True
cur = con.cursor()

def select(sql):
    return pd.read_sql(sql,con)
```

```python
sql = '''select t.* from winequality t order by t.alcohol desc'''
```

```python
df = select(sql)
```

```python
df.head()
```
