iterrows and iteritems are used to iterate over dataframes with a loop

https://www.youtube.com/watch?v=njeJ5WOmlJk

```
import pandas as pd
```

```
data = {
    "No" : [1, 2, 3],
    "Name" : ['Alex', 'Maria', 'John'],
    "Percentage" : [80, 85, 89]
}
df = pd.DataFrame(data, index = ['index1', 'index2', 'index3'])
df
```

## **Get column names**

```
for i in df:
  print(i)
```

```
df.columns
```

## **iterrows()**

```
# assuming the dataframe is df
for i in df.iterrows():
  print(i)
```

```
print(df)
print()
for (i,j) in df.iterrows():
  print(i)
  print(j)
```

## **iteritems()**

```
print(df)
print()
for i in df.iteritems():
  print(i)
```

```
print(df)
print()
for col,colSeries in df.iteritems():
  print(col)
  print(colSeries)
```

```
print(df)
print()
for col,colSeries in df.iteritems():
  print(col)
  for i in colSeries:
    print(i)
```

```
for i, row in df.iterrows():
    print(i, row[0], row[1], row[2])
```

```
# i shows the index
# As shown above, we do not have to give column names. The above code and this one give the same output.
for i, col in df.iterrows():
    print(i, col['No'], col['Name'], col['Percentage'])
```

## **List Comprehension**
get a list of one column

```
list_name = [i for i in df['Name']]
list_name
```

## **itertuples()**
Iterate Rows with Index and each Row as a List.

DO NOT try to change data in the df this way, but it is convenient for iterating.

Itertuples is supposed to be much faster than Iterrows for large datasets.

```
for row in df.itertuples(name=None): 
    print(row[0], row[1], row[2], row[3])
```
